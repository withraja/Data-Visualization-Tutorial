# **Bar** Chart

We will be **using a dataset about used cars**.  
That contains two columns; the first column is the car brand, and the second is the number of car advertisements of that brand.  
➡️ **Bar chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar

We need to **import pandas** for read the data **and matplotlib** for visualization 

````python
import pandas as pd
import matplotlib.pyplot as plt
````

Then we read the .csv file and named it **df_cars**

````python
df_cars = pd.read_csv("Bar_Chart_Dataset.csv")
df_cars
````

### **Basics** Parameter | x height plt.show

The basic parameters for the bar chart are **x** (x-axes) and **height** (y-axes)  
We need to always include **plt . show** to show our visualization

````python
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"])
plt.show()
````

### Solution for **Overlapping** Labels | figsize rotation

Default size of the figure is 6.4 x 4.8 inches  
Usually when you **use 9 x 6 size** will be appropriate for most visualization  

If you want to rotate the labels in the x axes, you can **use rotation in the plt . xticks**

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"])
plt.xticks(rotation = (45))  # rotate by 45 degrees
plt.show()
````

### Combine with **Seaborn**

Make sure you read the readme.md because there will be explanation why we use two library for visualization 👍

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # import seaborn
sns.set()              # set seaborn as the main theme
````

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"])
plt.xticks(rotation = (45))  # rotate by 45 degrees
plt.show()
````

### Customize **Colors** | color 

The use of **"color" is located in matplotlib**, but the library used for variation of colors is seaborn

We can give coloring **according to the number of existing columns** or one colours for all columns.  
It can be with a color name specified by the library (such as yellow, magenta etc.) or with hexcodes or etc.  
➡️ **Color Documentation** https://matplotlib.org/stable/gallery/color/color_demo.html#sphx-glr-gallery-color-color-demo-py

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"],
        color = {"r", "g", "b", "w", "m", "y", "c"})
plt.xticks(rotation = (45))
plt.show()
````

### Add **Labels** | title ylabels xlabels etc.

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"],
        color = {"r", "g", "b", "w", "m", "y", "c"})
plt.title("Car Listings by Brand", fontsize = 20, fontweight = "bold")  # Add Title and customize it
plt.ylabel("Number of Listings", fontsize = 15, labelpad = 20)                         # Add Y-Label and customize it
plt.xticks(rotation = (45))
plt.show()
````

### **Export** Chart | savefig

We need to specify the name and **file format**

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"],
        color = {"r", "g", "b", "w", "m", "y", "c"})
plt.title("Car Listings by Brand", fontsize = 20, fontweight = "bold")  # Add Title and customize it
plt.ylabel("Number of Listings", fontsize = 15, labelpad = 20)                         # Add Y-Label and customize it
plt.xticks(rotation = (45))
plt.savefig("Car Listings.png")
````

# **Pie** Chart

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_fuel_type = pd.read_csv("Pie_Chart_Dataset.csv")
df_fuel_type
````

### **Add Labels** in Pie Chart | labels autopct

Pie charts usually display data in percentage form. Hence **autopct (auto-percentage) is always used**

In the following autopct we set the value to float with 2 decimal places ```.2f```  
Added with the percent symbol as a string ```%```  
Then **the first and the last "%" symbol indicates** the text given in autopct argument as a value that should be considered as the active math-text formatting and follow the rules for that math-text formatting system

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%")
plt.show()
````

### **Customize** the Labels and **Add Legend** | textprops plt.legend

We can put all our customizations for labels in multiple dictionaries to save on writing code.  
The code that holds the dictionary is **textprops**

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%",
        textprops = {"size": "large",     # textprops
                     "fontweight": "bold",
                     "color": "white"})
plt.legend()                                # add legend
plt.title("Cars by Engine Fuel Type",       # add title
          fontsize = "20", 
          fontweight = "bold")       
plt.show()
````

### **Combine with Seaborn** Color Pallete

````python
sns.set_palette("colorblind")
````

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%",
        textprops = {"size": "large",     # textprops
                     "fontweight": "bold",
                     "color": "white"})
plt.legend()                                # add legend
plt.title("Cars by Engine Fuel Type",       # add title
          fontsize = "20", 
          fontweight = "bold")       
plt.show()
````
