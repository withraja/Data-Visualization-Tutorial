# **Bar** Chart

We will be **using a dataset about used cars**.  
That contains two columns; the first column is the car brand, and the second is the number of car advertisements of that brand.  
➡️ **Bar chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar

We need to **import pandas** for read the data **and matplotlib** for visualization 

````python
import pandas as pd
import matplotlib.pyplot as plt
````

Then we read the .csv file and named it **df_cars**

````python
df_cars = pd.read_csv("Bar_Chart_Dataset.csv")
df_cars
````

### **Basics** Parameter | x height plt.show

The basic parameters for the bar chart are **x** ( x-axes ) and **height** ( y-axes )  
We always need to include **plt . show** to show our visualization

````python
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"])
plt.show()
````

### Solution for **Overlapping** Labels | figsize rotation

The default size of the figure is 6.4 x 4.8 inches  
Usually when you **use 9 x 6 size** will be appropriate for most visualization  

If you want to rotate the labels in the x-axes, you can **use rotation in the plt . xticks**

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"])
plt.xticks(rotation = (45))  # rotate by 45 degrees
plt.show()
````

### Combine with **Seaborn**

Make sure you **read the readme.md** because there will be an explanation of why we use two libraries for visualization 👍

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # import seaborn
sns.set()              # set seaborn as the main theme
````

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"])
plt.xticks(rotation = (45))  # rotate by 45 degrees
plt.show()
````

### Customize **Colors** | color 

We can give colouring **according to the number of existing columns** or one colour for all columns  
It can be with a colour name specified by the library (such as yellow, magenta etc.) or with hex codes, etc  

➡️ **Color Documentation** https://matplotlib.org/stable/gallery/color/color_demo.html#sphx-glr-gallery-color-color-demo-py

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"],
        color = {"r", "g", "b", "w", "m", "y", "c"})
plt.xticks(rotation = (45))
plt.show()
````

### Add **Labels** | title ylabels xlabels etc.

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"],
        color = {"r", "g", "b", "w", "m", "y", "c"})
plt.title("Car Listings by Brand", fontsize = 20, fontweight = "bold")  # Add Title and customize it
plt.ylabel("Number of Listings", fontsize = 15, labelpad = 20)                         # Add Y-Label and customize it
plt.xticks(rotation = (45))
plt.show()
````

### **Export** Chart | savefig

We need to specify the name and **file format**

````python
plt.figure(figsize = (9,6))
plt.bar(x = df_cars["Brand"], 
        height = df_cars["Cars Listings"],
        color = {"r", "g", "b", "w", "m", "y", "c"})
plt.title("Car Listings by Brand", fontsize = 20, fontweight = "bold")  # Add Title and customize it
plt.ylabel("Number of Listings", fontsize = 15, labelpad = 20)                         # Add Y-Label and customize it
plt.xticks(rotation = (45))
plt.savefig("Car Listings.png")
````

# **Pie** Chart

We will be **using a dataset about engine fuel type**.  
That contains two columns; the first column is engine type, and the second is the number of cars that used that engine.  
➡️ **Pie chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_fuel_type = pd.read_csv("Pie_Chart_Dataset.csv")
df_fuel_type
````

### **Add Labels** in Pie Chart | labels autopct

Pie charts usually display data in percentage form. Hence **autopct (auto-percentage) is always used**

In the following autopct we set the value to float with 2 decimal places ```.2f```  
Added with the percent symbol as a string ```%```  
Then **the first and the last "%" symbol indicates** the text given in autopct argument as a value that should be considered as the active math-text formatting and follow the rules for that math-text formatting system

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%")
plt.show()
````

### **Customize** the Labels and **Add Legend** | textprops plt.legend

We can label all customisations in one dictionary to save on writing code  
The code that holds the dictionary is **textprops**

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%",
        textprops = {"size": "large",     # textprops
                     "fontweight": "bold",
                     "color": "white"})
plt.legend()                                # add legend
plt.title("Cars by Engine Fuel Type",       # add title
          fontsize = "20", 
          fontweight = "bold")       
plt.show()
````

### **Combine with Seaborn** Color Pallete

````python
sns.set_palette("colorblind")
````

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%",
        textprops = {"size": "large",     # textprops
                     "fontweight": "bold",
                     "color": "white"})
plt.legend()                                # add legend
plt.title("Cars by Engine Fuel Type",       # add title
          fontsize = "20", 
          fontweight = "bold")       
plt.show()
````

# **Stacked Area** Chart

➡️ **How to Read Stacked Area Chart** https://chartio.com/learn/charts/area-chart-complete-guide/

We will be **using a dataset about popularity of engine fuel type across the year**.  
It contains five columns; the first column is the year, the second column and so on is the type of fuel used by the engine. Such as diesel, petrol and others.
  
➡️ **Stack Area chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stackplot.html#matplotlib.pyplot.stackplot

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_fuel_engine_type = pd.read_csv("Stacked_Area_Chart_Dataset.csv")
df_fuel_engine_type.head()
````

````python
plt.stackplot(df_fuel_engine_type["Year"],     
              df_fuel_engine_type["Gas"],      
              df_fuel_engine_type["Petrol"],
              df_fuel_engine_type["Diesel"],
              df_fuel_engine_type["Other"])
plt.show()
````

### **Customize** the Chart | sns.**set_style** sns.**despine** edgecolor

````python
sns.set_style("white")                         # make the background white
colors = sns.color_palette("bright")           # we are using seaborn color palette
labels = ["Diesel", "Petrol", "Gas", "Other"]  # using a label for the legend, but it must be sequential according to the order of the variables used in the stackplot
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_type["Year"],     
              df_fuel_engine_type["Diesel"],      
              df_fuel_engine_type["Petrol"],
              df_fuel_engine_type["Gas"],
              colors = colors,
              edgecolor = "none")               # removing white edgecolor from the graph
plt.legend(labels = labels,                     # using labels as legend    
           loc = "upper left")                  # specify the position of the legend so that it is in the upper left corner
plt.ylabel("Number of Cars", fontsize = 16, labelpad = 20)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 20, weight = "bold", pad = 10)
sns.despine()                                   # to remove borders from a Seaborn plot
plt.show()
````

# **Line** Chart

We will be **using a dataset about stock market returns in the USA and UK**.  
It contains three columns; the first column is the **time-series**, the second column is **S&P 500 Index** which stands for USA Stock Market, and the third one is **FTSE 100** which stands for UK Stock Market.
  
➡️ **Line chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stackplot.html#matplotlib.pyplot.plot

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_stock_market = pd.read_csv("Line_Chart_Dataset.csv")
df_stock_market.head()
````

### Change **Datatype Format to Date** | to_datetime

When working with time-series data, it is crucial to have dates in the correct format to avoid complications in analyzing and plotting the data. So, **let's check the data type in the first column**.

````python
df_stock_market["Date"].dtypes
````

The first column dtype is **O** (stands for Object, String, or mixed), so **we need to convert it to date format**

````python
df_stock_market["New_Date"] = pd.to_datetime(df_stock_market["Date"])
df_stock_market["New_Date"].dtypes
````

Now the first column dtype is **< M8 which stands for date datatype in Pandas** 

### Create **Visualization** | plt.**plot**

There are various ways to plot multiple sets of data in line chart  
The most straight forward way is just to **call plot multiple times**

````python
labels = ["S&P 500", "FTSE 100"]                    
plt.figure(figsize = (20, 8))
plt.plot(df_stock_market["New_Date"], df_stock_market["GSPC500"])
plt.plot(df_stock_market["New_Date"], df_stock_market["FTSE100"])

# chart formatting elements: title, labels and legend
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize = 20, fontweight = "bold", pad = 15)
plt.ylabel("Returns", fontsize = 15, fontweight = "bold")
plt.xlabel("Date", fontsize = 15, fontweight = "bold")
plt.legend(labels = labels, fontsize = "large")
plt.show()
````

It's impossible to gain spesific insights in this chart as we have to many point in small surfaces  
We need to increase granularity of our chart by **choose a smaller time frame**

### **Slice** Our Chart

**We intend to highlight the end of 2008 timeframes** when the phenomenon of the Great Recession happened  

➡️ **The Great Recession** https://corporatefinanceinstitute.com/resources/economics/2008-2009-global-financial-crisis/

````python
df_stock_market_sliced = df_stock_market[(df_stock_market.New_Date >= '2007-09-01') &
                                        (df_stock_market.New_Date <= '2008-12-31')]
````

````python
sns.set_style("white")
labels = ["S&P 500", "FTSE 100"]                    
plt.figure(figsize = (20, 8))
plt.plot(df_stock_market_sliced["New_Date"], df_stock_market_sliced["GSPC500"], linewidth = 3)
plt.plot(df_stock_market_sliced["New_Date"], df_stock_market_sliced["FTSE100"], linewidth = 3)

# chart formatting elements: title, labels and legend
plt.title("S&P vs FTSE Returns (September - December 2008)", fontsize = 20, fontweight = "bold", pad = 15)
plt.ylabel("Returns", fontsize = 15, fontweight = "bold")
plt.xlabel("Date", fontsize = 15, fontweight = "bold")
plt.legend(labels = labels, fontsize = "large")
sns.despine()
plt.show()
````

### **Interpretation** 💡

Up to August 2008, the return is more or less stable. Of course, there is movement, but not too drastic.  
However, **after this point, we see enormous swings in our data caused by the crash**. Though it occurs in the US, the UK Stock Market has the same pattern.  

We could conclude that the effect is global partly because the US is an economic superpower or because all significant financial systems worldwide are connected.
If in the past, one country's economic downturn meant another's uprising.  
Nowadays, the economic downturn of one big country means the downturn of the economic of the whole world.

# **Histogram**

We will be **using a dataset about real estate property listings in California**.  
Each listings is described through various features, starting with ID. But we will concentrate on the **Price column**  

➡️ **Histogram documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_real_estate = pd.read_csv("Histogram_Data.csv")
df_real_estate.head()
````

### Create **Visualization** | plt.**hist**

````python
sns.set_style("white") 
plt.figure(figsize = (8, 6)) 
plt.hist(df_real_estate["Price"],    # the variable on which to create the histogram
         bins = 8,                   # create a histogram with 8 bins
         color = "#108A99")
plt.title("Distribution of Real Estate Prices", fontsize = 18, weight = "bold", pad = 10)
plt.xlabel("Price in (000's $)", fontsize = 15, weight = "bold", labelpad = 10)
plt.ylabel("Number of Properties", fontsize = 15, weight = "bold", labelpad = 10)
sns.despine()
plt.show()
````
