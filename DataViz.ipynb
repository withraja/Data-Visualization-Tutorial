# **Bar** Chart

We will be **using a dataset about used cars**.  
That contains two columns; the first column is the car brand, and the second is the number of car advertisements of that brand.

The basic parameters for the bar chart are **x** ( x-axes ) and **height** ( y-axes )

# **Pie** Chart

We will be **using a dataset about engine fuel type**.  
That contains two columns; the first column is engine type, and the second is the number of cars that used that engine.  

➡️ **Pie chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_fuel_type = pd.read_csv("Pie_Chart_Dataset.csv")
df_fuel_type
````

### **Add Labels** in Pie Chart | labels autopct

Pie charts usually display data in percentage form. Hence **autopct (auto-percentage) is always used**

In the following autopct we set the value to float with 2 decimal places ```.2f```  
Added with the percent symbol as a string ```%```  
Then **the first and the last "%" symbol indicates** the text given in autopct argument as a value that should be considered as the active math-text formatting and follow the rules for that math-text formatting system

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%")
plt.show()
````

### **Customize** the Labels and **Add Legend** | textprops plt.legend

We can label all customisations in one dictionary to save on writing code  
The code that holds the dictionary is **textprops**

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%",
        textprops = {"size": "large",     # textprops
                     "fontweight": "bold",
                     "color": "white"})
plt.legend()                                # add legend
plt.title("Cars by Engine Fuel Type",       # add title
          fontsize = "20", 
          fontweight = "bold")       
plt.show()
````

### **Combine with Seaborn** Color Pallete

````python
sns.set_palette("colorblind")
````

````python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_type["Number of Cars"],
        labels = df_fuel_type["Engine Fuel Type"].values,
        autopct = "%.2f%%",
        textprops = {"size": "large",     # textprops
                     "fontweight": "bold",
                     "color": "white"})
plt.legend()                                # add legend
plt.title("Cars by Engine Fuel Type",       # add title
          fontsize = "20", 
          fontweight = "bold")       
plt.show()
````

## Interpretation 💡

We can see that the **predominant engines types are diesel by 46%** followed by petrol with roughly 36%  
1/6 of the cars have a gas tank, when it's only about 4% drove on some other type of fuel  
After all not much interpretation is necessary and that's why Pie charts are often the prefer choice for non-technical audience 

# **Stacked Area** Chart

➡️ **How to Read Stacked Area Chart** https://chartio.com/learn/charts/area-chart-complete-guide/

We will be **using a dataset about popularity of engine fuel type across the year**.  
It contains five columns; the first column is the year, the second column and so on is the type of fuel used by the engine. Such as diesel, petrol and others.
  
➡️ **Stack Area chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stackplot.html#matplotlib.pyplot.stackplot

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_fuel_engine_type = pd.read_csv("Stacked_Area_Chart_Dataset.csv")
df_fuel_engine_type.head()
````

````python
plt.stackplot(df_fuel_engine_type["Year"],     
              df_fuel_engine_type["Gas"],      
              df_fuel_engine_type["Petrol"],
              df_fuel_engine_type["Diesel"],
              df_fuel_engine_type["Other"])
plt.show()
````

### **Customize** the Chart | sns.**set_style** sns.**despine** edgecolor

````python
sns.set_style("white")                         # make the background white
colors = sns.color_palette("bright")           # we are using seaborn color palette
labels = ["Diesel", "Petrol", "Gas", "Other"]  # using a label for the legend, but it must be sequential according to the order of the variables used in the stackplot
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_type["Year"],     
              df_fuel_engine_type["Diesel"],      
              df_fuel_engine_type["Petrol"],
              df_fuel_engine_type["Gas"],
              colors = colors,
              edgecolor = "none")               # removing white edgecolor from the graph
plt.legend(labels = labels,                     # using labels as legend    
           loc = "upper left")                  # specify the position of the legend so that it is in the upper left corner
plt.ylabel("Number of Cars", fontsize = 16, labelpad = 20)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 20, weight = "bold", pad = 10)
sns.despine()                                   # to remove borders from a Seaborn plot
plt.show()
````

## Interpretation 💡

From our chart, we can see that **Petrol and Diesel well pretty much the preferred fuel type**  

### Distribution of the Dataset
There are few cars from the '80s in our sample; many vehicles did not make it to the listing because they broke, and automobiles were generally less common  
As time passed, Diesel and Petrol were playing catch for the leading fuel type  
Most notably, **in 1997 Diesel became the preferred engine fuel type**  
Then followed the early 2000s with an explosion of Diesel cars being produced  
From then onward, it's clear that Diesel has dominated the market  
After this time frame, many countries start phasing out diesel engine types, and we expect them to be entirely gone by 2040  

### Gas-fueled cars 
Gas cars had their heyday during the 2007 to 2009 period  
According to our chart, we can also see a very pronounced dip in the number of used vehicles bought in 2008 and 2009   This is most definitely related to the stock market crisis during the period. Mainly because we can observe the same trend across all categories  
The drop in 2016 is simply because we didn't have it that year

In our chart, we aided the interpretation by ordering the categories by size  
We also choose a contrasting colour for the smallest category "Gas", which allows us to observe its development and spot an emerging trend. 

# **Line** Chart

We will be **using a dataset about stock market returns in the USA and UK**.  
It contains three columns; the first column is the **time-series**, the second column is **S&P 500 Index** which stands for USA Stock Market, and the third one is **FTSE 100** which stands for UK Stock Market.
  
➡️ **Line chart documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stackplot.html#matplotlib.pyplot.plot

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_stock_market = pd.read_csv("Line_Chart_Dataset.csv")
df_stock_market.head()
````

### Change **Datatype Format to Date** | to_datetime

When working with time-series data, it is crucial to have dates in the correct format to avoid complications in analyzing and plotting the data. So, **let's check the data type in the first column**.

````python
df_stock_market["Date"].dtypes
````

The first column dtype is **O** (stands for Object, String, or mixed), so **we need to convert it to date format**

````python
df_stock_market["New_Date"] = pd.to_datetime(df_stock_market["Date"])
df_stock_market["New_Date"].dtypes
````

Now the first column dtype is **< M8 which stands for date datatype in Pandas** 

### Create **Visualization** | plt.**plot**

There are various ways to plot multiple sets of data in line chart  
The most straight forward way is just to **call plot multiple times**

````python
labels = ["S&P 500", "FTSE 100"]                    
plt.figure(figsize = (20, 8))
plt.plot(df_stock_market["New_Date"], df_stock_market["GSPC500"])
plt.plot(df_stock_market["New_Date"], df_stock_market["FTSE100"])

# chart formatting elements: title, labels and legend
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize = 20, fontweight = "bold", pad = 15)
plt.ylabel("Returns", fontsize = 15, fontweight = "bold")
plt.xlabel("Date", fontsize = 15, fontweight = "bold")
plt.legend(labels = labels, fontsize = "large")
plt.show()
````

It's impossible to gain spesific insights in this chart as we have to many point in small surfaces  
We need to increase granularity of our chart by **choose a smaller time frame**

### **Slice** Our Chart

**We intend to highlight the end of 2008 timeframes** when the phenomenon of the Great Recession happened  

➡️ **The Great Recession** https://corporatefinanceinstitute.com/resources/economics/2008-2009-global-financial-crisis/

````python
df_stock_market_sliced = df_stock_market[(df_stock_market.New_Date >= '2007-09-01') &
                                        (df_stock_market.New_Date <= '2008-12-31')]
````

````python
sns.set_style("white")
labels = ["S&P 500", "FTSE 100"]                    
plt.figure(figsize = (20, 8))
plt.plot(df_stock_market_sliced["New_Date"], df_stock_market_sliced["GSPC500"], linewidth = 3)
plt.plot(df_stock_market_sliced["New_Date"], df_stock_market_sliced["FTSE100"], linewidth = 3)

# chart formatting elements: title, labels and legend
plt.title("S&P vs FTSE Returns (September - December 2008)", fontsize = 20, fontweight = "bold", pad = 15)
plt.ylabel("Returns", fontsize = 15, fontweight = "bold")
plt.xlabel("Date", fontsize = 15, fontweight = "bold")
plt.legend(labels = labels, fontsize = "large", loc = "upper left")
sns.despine()
plt.show()
````

### **Interpretation** 💡

Up to August 2008, the return is more or less stable. Of course, there is movement, but not too drastic.  
However, **after this point, we see enormous swings in our data caused by the crash**. Though it occurs in the US, the UK Stock Market has the same pattern.  

We could conclude that the effect is global partly because the US is an economic superpower or because all significant financial systems worldwide are connected.
If in the past, one country's economic downturn meant another's uprising.  
Nowadays, the economic downturn of one big country means the downturn of the economic of the whole world.

# **Scatter** Plot

We will use the same dataset as the Histogram tutorial **about real estate property listings in California**.  
This tutorial will examine the **correlation between the Price and the Area Squares**.  

➡️ **Scatterplot documentation** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
df_scatter = pd.read_csv("Scatterplot_Data.csv")
df_scatter.head()
````

### Create Visualization **with Matplotlib** | plt.**scatter**

````python
plt.figure(figsize = (12, 8)) 
sns.set_style("white")
scatter = plt.scatter(df_scatter['Area (ft.)'], # x-axis
            df_scatter['Price'],                # y-axis
            alpha = 0.6,      # transparency level of points on the plot. Used to avoid overplotting
            c = df_scatter['Building Type'],    # "c" stands for color, it must be an array with integer values or a list of color
            cmap = 'viridis')  # color pallete in the matplotlib library

# Formatting additional elements, such as legends, titles and axis labels
plt.legend(*scatter.legend_elements(),  # *asterisk means "unpacking" --> unpack the scatter variable
                                        # legend_elements automatically determine a useful number of legend entries to be shown. It will search for "c" and...
                                        # search for how many differences in the "color" values
           loc = "upper left",
           title = "Building Type")
plt.title("Relationship between Area and Price of California Real Estate", 
          fontsize = 18, 
          weight = "bold",
          pad = 20)
plt.xlabel("Area (ft.)", weight = "bold", fontsize = 14, labelpad = 10)
plt.ylabel("Price (000's of $)", weight = "bold", fontsize = 14, labelpad = 10)
sns.despine()
plt.show()
````

# **Regression** Plot

We will be **using a dataset about company advertising budget VS sales** across 200 different markets  
That contains two columns; the first column is **Ad Budget** in thousand of $, and the second is the number of sales thousand of units  

➡️ **Regression line documentation** https://seaborn.pydata.org/generated/seaborn.regplot.html

````python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
````

````python
df_ad_expenditure = pd.read_csv("Regression_Plot.csv")
df_ad_expenditure.head()
````

Seaborn's  `regplot()`  function is **specifically designed for creating scatter plots with regression lines**, and it provides a number of useful features, such as confidence intervals and options for different types of regression models  

So in this tutorial, we will rely on seaborn library

### sns.**regplot**

````python
plt.figure(figsize = (10, 8))
sns.regplot(x = "Budget",                   # x-axis
            y = "Sales",                    # y-axis
            data = df_ad_expenditure,       # dataset
            scatter_kws = {'color': 'black'},   # color for the points
            line_kws = {'color': 'red'})    # color for the regression line
plt.xlabel("Ad Expenditure in (000's $)", weight = "bold", fontsize = 14, labelpad = 10)
plt.ylabel("Sales in (000's Units)", weight = "bold", fontsize = 14, labelpad = 10)
plt.title("Effect of Ad Expenditure on Sales", fontsize = 18, weight = "bold", pad = 15)
plt.show()
````

A **95% confidence interval will be added by default**. It's the gray area around the regression line   
Or we can specify it with `ci` argument in seaborn code

### sns.**lmplot**

Another functions to create regression line is lmplot.  
The `lmplot()` is a more versatile function that can be **used to create more complex plots with groupings and multiple regression lines**. If you only need a single regression plot and prefer to work with individual x and y variables, regplot is the better choice.  

➡️ **lmplot documentation** https://seaborn.pydata.org/generated/seaborn.lmplot.html

````python
sns.lmplot(x = "Budget", 
           y = "Sales", 
           data = df_ad_expenditure, 
           height = 8,                 # height and width of the plot
           scatter_kws = {'color': 'k'}, 
           line_kws = {'color': 'red'})  
plt.xlabel("Ad Expenditure in (000's $)", weight = "bold", fontsize = 14, labelpad = 10)
plt.ylabel("Sales in (000's Units)", weight = "bold", fontsize = 14, labelpad = 10)
plt.title("Effect of Ad Expenditure on Sales", fontsize = 18, weight = "bold", pad = 15)
plt.show()
````

**Because we are only comparing two variables** (without any particular grouping or conditional), the visuals displayed by regplot and lmplot are identical

# Combo Chart | **Bar and Line**

We will be **using a dataset about annual user by KDNuggets** spesifically about numbers of Python users across the years    
That contains three columns; the first column is **Years**, the second is **the portion of python users in decimals**, and the last column is **the number of survey participants**  

➡️ **How to make dual axis in Python** https://towardsdatascience.com/creating-a-dual-axis-combo-chart-in-python-52624b187834

````python
import pandas as pd
import matplotlib.pyplot as plt 

from matplotlib.ticker import PercentFormatter  # Converts values into percentage format, we will need it later

import seaborn as sns 
sns.set()
````

````python
df_kdnuggets = pd.read_csv("Bar_Line_Chart_Data.csv")
df_kdnuggets.head()
````

Step by step. First of all, **you need to understand the different between "figures" and "axes"**  
The figures is the complete figure we're creating in matplotlib. It includes all chart and all axes  
The axes is the actual plot 

### **Bar** Chart | Number of Participants

We named our variable as **ax**

````python
# Determine the size of our visualization
fig, ax = plt.subplots(figsize = (10, 7))

ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],
       color = "#6A2C70")

ax.set_ylabel("Number of Participants", 
              weight = "bold",
              fontsize = 14,
              labelpad = 10)

ax.tick_params(axis = "y", 
               width = 2, 
               labelsize = "large")
````

### **Line** Chart | Python Users

````python
# Determine the size of our visualization
fig, ax = plt.subplots(figsize = (10, 7))

# Changing the y-axis to display percentages on a scale from 0% to 100% with PercentFormatter
# Of course with 0 minimum and 1 maximum
ax.set_ylim(0, 1)
ax.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))

# Creating and styling the line chart
ax.plot(df_kdnuggets["Year"],
         df_kdnuggets["Python Users"], 
         color = "#F08A5D", 
         marker = "o")

ax.set_ylabel("Python Users", 
               weight = "bold",
               fontsize = 14,
               labelpad = 10)

ax.tick_params(axis = "y", 
                width = 2, 
                labelsize = "large")
````

### Combine with **twinx ( )**

We will combine all codes from two plot (bar and line) with twinx  
We will also **rename our line chart from ax > ax1**

````python
sns.set_style("white")
fig, ax = plt.subplots(figsize = (10, 7))

# BAR CHART
ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],
       color = "#6A2C70")
ax.set_ylabel("Number of Participants", 
              weight = "bold",
              fontsize = 14,
              labelpad = 10)
ax.tick_params(axis = "y", 
               width = 2, 
               labelsize = "large")

# COMBINE BAR AND LINE CHART (making line chart as secondary y-axis)
ax1 = ax.twinx()

# LINE CHART
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets["Year"],
         df_kdnuggets["Python Users"], 
         color = "#F08A5D", 
         marker = "o")
ax1.set_ylabel("Python Users", 
               weight = "bold",
               fontsize = 14,
               labelpad = 10)
ax1.tick_params(axis = "y", 
                width = 2, 
                labelsize = "large")

# IMPROVEMENT
ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)", fontsize = "20", weight = "bold", pad = 15)
plt.show()
````
